```python
import pandas as pd
```

```python
# Especifica la ruta de tu archivo Excel
ruta_excel = 'data/cdro_G8.xlsx'

# Carga el archivo Excel
archivo_excel = pd.ExcelFile(ruta_excel)

# Extrae los nombres de todas las hojas
nombres_de_pestañas = archivo_excel.sheet_names

# Imprime los nombres de todas las hojas
print(nombres_de_pestañas)
```

```python
# Lee la hoja específica del archivo Excel y carga los datos en un DataFrame
nombre_hoja = 'nombre_de_tu_hoja'
df = pd.read_excel(ruta_excel, sheet_name=nombre_hoja)
```

```python
# Itera sobre los nombres de las hojas y carga cada una en un DataFrame
dataframes = {}
for nombres_de_pestañas in archivo_excel.sheet_names:
    df = pd.read_excel(archivo_excel, sheet_name=nombres_de_pestañas, skiprows=4, usecols=lambda x: 'Unnamed' not in x)  # Comienza desde la fila 5
    df = df.iloc[3:]  # Elimina las primeras 3 filas
    dataframes[nombres_de_pestañas] = df
    # Encuentra la última fila que contiene la palabra "TOTAL" en la primera columna
    indice_total = df[df.iloc[:, 0].str.contains('TOTAL', na=False)].index.max()

    # Si no encuentras la palabra "TOTAL", el índice será NaN, lo cual debería manejarse
    if pd.isnull(indice_total):
        print("No se encontró la palabra 'TOTAL' en la primera columna.")
    
    else:
    # Elimina todas las filas debajo de la última fila que contiene la palabra "TOTAL"
        df = df.iloc[:indice_total + 1]
```
